## Word Cloud for Video Titles

```python
# Python program to generate WordCloud
 
# importing all necessary modules
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
 

comment_words = ''
stopwords = set(STOPWORDS)
 
# iterate through the csv file
for val in df1_video['Video title']:
     
    # typecaste each val to string
    val = str(val)
 
    # split the value
    tokens = val.split()
     
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
     
    comment_words += " ".join(tokens)+" "
 
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
 
# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.title('Word Cloud of All Video Titles')
plt.show()
```

### Video Data observed by country dimension

```python
df2.columns
```

```python
df2['Likes'] = df2['Video Likes Added'] - df2['Video Likes Removed']
df2['Subs'] = df2['User Subscriptions Added'] - df2['User Subscriptions Removed']
```

```python
top10_views_df = df2[df2['Country Code'].isin(list(df2.groupby('Country Code')['Views'].sum().sort_values(ascending = False)[:10].index))]
```

```python
country_df = top10_views_df[['Country Code','Views', 'Average View Percentage', 'Average Watch Time', 'Likes', 'Subs']]
```

```python
country_df.groupby('Country Code')[['Views']].sum().plot(kind = 'bar')
plt.title('Views bar plot from countries having top 10 views')
plt.show()
country_df.groupby('Country Code')[[ 'Likes', 'Subs']].sum().plot(kind = 'bar')
plt.title('Likes/Subscribers bar plot from countries having top 10 likes/subscribers')
plt.show()
country_df.groupby('Country Code')[['Average Watch Time']].sum().plot(kind = 'bar')
plt.title('Average Watch Time bar plot from countries having top 10 Average Watch Time')
plt.show()
```

```python
top5_views_df = df2[df2['Country Code'].isin(list(df2.groupby('Country Code')['Views'].sum().sort_values(ascending = False)[:5].index))]
```

```python
subs_top5 = top5_views_df[top5_views_df['Is Subscribed'] == True]
no_subs_top5 = top5_views_df[top5_views_df['Is Subscribed'] == False]
```

```python
subs_df = subs_top5[['Country Code','Views', 'Average View Percentage', 'Average Watch Time', 'Likes', 'Subs']].groupby('Country Code').sum()
subs_df['subs'] = True
no_subs_df = no_subs_top5[['Country Code','Views', 'Average View Percentage', 'Average Watch Time', 'Likes', 'Subs']].groupby('Country Code').sum()
no_subs_df['subs'] = False
```

```python
final_subs_df = pd.merge(subs_df, no_subs_df, on = 'Country Code')
```

```python
final_subs_df[['Views_x', 'Views_y']].plot(kind = 'bar')
plt.legend(['Subscribed', 'No Subscribed'])
plt.title('Views Data of Top 5 Views country(subcribed/no subscribed)')
plt.show()
final_subs_df[['Likes_x', 'Likes_y']].plot(kind = 'bar')
plt.legend(['Subscribed', 'No Subscribed'])
plt.title('Likes Data of Top 5 Views country(subcribed/no subscribed)')
plt.show()
final_subs_df[['Average Watch Time_x', 'Average Watch Time_y']].plot(kind = 'bar')
plt.legend(['Subscribed', 'No Subscribed'])
plt.title('Average Watch Time Data of Top 5 Views country(subcribed/no subscribed)')
plt.show()
```

```python
final_subs_df
```

```python
by_country = df2.groupby(['Country Code']).sum()
by_country['Likes'] = by_country['Video Likes Added'] - by_country['Video Likes Removed']
by_country['Subs'] = by_country['User Subscriptions Added'] - by_country['User Subscriptions Removed']
by_country_sub = by_country.loc[['US', 'IN', 'GB', 'CA', 'DE', 'BR'],['Views','Likes','Subs']]
pop = zip(['US', 'IN', 'GB', 'CA', 'DE', 'BR'], [331002651, 1380004385, 67886011, 37742154, 83783942, 212559417])
country_pop = pd.DataFrame(pop, columns = ['Country', 'Population'])
country_pop = country_pop.set_index('Country')
country_agg = country_pop.merge(by_country_sub, how='inner', left_index = True, right_index = True)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize = (10, 3), tight_layout = True)

country_agg['Population'].plot(kind='bar', width = 0.5, ax=ax[0], color = '#d62728')
ax[0].set_title('Population')
ax[0].set_yticks(np.linspace(0, 1500000000, 7))
ax[0].set_yticklabels([str(x)+'B' for x in np.linspace(0, 1.5, 7)])

country_agg['Views'].plot(kind='bar', width = 0.5, ax=ax[1], color = '#2ca02c')
ax[1].set_title('Number of Views')
ax[1].set_yticks(np.linspace(0, 1500000, 7))
ax[1].set_yticklabels([str(x)+'M' for x in np.linspace(0, 1.5, 7)])

country_agg.iloc[:,2:4].plot(kind='bar', width = 0.5, ax=ax[2])
ax[2].set_title('Number of Likes/Subs')
ax[2].set_yticks(np.linspace(0, 60000, 7))
ax[2].set_yticklabels([str(x)+'K' for x in np.linspace(0, 60, 7)])
ax[2].legend()

fig.autofmt_xdate(rotation = 0)

plt.show()
```

```python
for col, val in country_agg.iteritems():
    if (col != 'Population'):
        country_agg[col+'_per_Capita'] = val/country_agg['Population']
per_cap = country_agg.loc[:,country_agg.columns.str.contains('per_Capita')]
per_cap.columns = ['Views', 'Likes', 'Subs']
fig, ax = plt.subplots(nrows=1, ncols=2, figsize = (9, 4), tight_layout = True)

per_cap['Views'].plot(kind='bar', width = 0.5, ax=ax[0], color = '#2ca02c')
ax[0].set_title('Per Capita Views')
#ax[0].set_yticks(np.linspace(0, 1500000, 7))
#ax[0].set_yticklabels([str(x)+'M' for x in np.linspace(0, 1.5, 7)])

per_cap[['Likes', 'Subs']].plot(kind='bar', width = 0.5, ax=ax[1])
ax[1].set_title('Per Capita Likes/Subs')
#ax[1].set_yticks(np.linspace(0, 60000, 7))
#ax[1].set_yticklabels([str(x)+'K' for x in np.linspace(0, 60, 7)])
ax[1].legend()

fig.autofmt_xdate(rotation = 0)

plt.show()
```

```python
f = plt.figure()

ax1 = f.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
ax2 = f.add_axes([0.47, 0.31, 0.42, 0.34]) # zoom

# Main figure Axes 
per_cap.T.plot(kind = 'bar', ax = ax1)
ax1.set_title('Views/Likes/Subs by Top Countries')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=0 )
ax1.legend(ncol=3)

# Insert figure Axes 
per_cap.iloc[:,1:].T.plot(kind = 'bar', ax = ax2)
ax2.set_title('Zoom on Likes/Subs')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=0 )
ax2.get_legend().remove()

plt.show()
```
